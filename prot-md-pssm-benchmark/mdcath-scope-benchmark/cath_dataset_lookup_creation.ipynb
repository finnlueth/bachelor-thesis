```python
import pandas as pd

# Read the CATH domain list file
cath_df = pd.read_csv(
    "data/cath-domain-list-S100-v4_2_0.txt",
    comment="#",
    delim_whitespace=True,
    names=[
        "domain_id",
        "class_id",
        "arch_id",
        "top_id",
        "hom_id",
        "S35",
        "S60",
        "S95",
        "S100",
        "S100_count",
        "length",
        "resolution",
    ],
)
```

```python
cath_df
```

```python
mdcath_domains = set()
with open("data/mdCATH_sequence_AA.fasta") as f:
    for line in f:
        if line.startswith(">"):
            domain_id = line[1:].split()[0]
            mdcath_domains.add(domain_id)

print(f"Number of domains in mdCATH: {len(mdcath_domains)}")
```

```python
len(mdcath_domains)
```

```python
cath_df = cath_df[cath_df["domain_id"].isin(mdcath_domains)]

cath_df.loc[:, "cath_code"] = cath_df.apply(
    lambda x: f"{x['class_id']}.{x['arch_id']}.{x['top_id']}.{x['hom_id']}", axis=1
)

display(cath_df)
```

```python
cath_df[["domain_id", "cath_code"]].to_csv(
    "data/mdCATH_lookup.fix.tsv", sep="\t", index=False
)
```
