```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


def plot_pssm_heatmap(path):
    df = pd.read_csv(
        path,
        skiprows=1,
        delim_whitespace=True,
    )

    plt.figure(figsize=(15, 10))

    sns.heatmap(
        df,
        cmap="RdYlBu_r",
        vmin=0,
        vmax=1,
        xticklabels=True,
        yticklabels=False,
    )

    protein_name = path.split("/")[-1].split(".")[0]
    plt.title(f"PSSM Heatmap for {protein_name}")
    plt.xlabel("Amino Acids")
    plt.ylabel(f"Sequence Position: length {len(df)}")

    plt.tight_layout()
    plt.show()
```

```python
item = "d2idra_"
print("New")
path = f"../tmp/data/generated_pssms/scope40_prot-md-pssm-2025-04-02-15-20-18-dataset_320_0_prostt5/{item}.tsv"
plot_pssm_heatmap(path)
print("Old")
path = f"../tmp/data/generated_pssms/scope40_prot-md-pssm-2025-03-05-17-43-47-full-dataset/{item}.tsv"
plot_pssm_heatmap(path)
```
